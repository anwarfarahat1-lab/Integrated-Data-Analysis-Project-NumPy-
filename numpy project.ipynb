```python
import numpy as np
```

```python
data = [
[25,70,120,180],
[30,80,130,190],
[22,65,110,170],
[45,90,150,210],
[50,85,145,200],
[28,75,125,185],
[35,88,140,195],
[40,95,155,220],
[32,78,135,188],
[27,72,122,182]
]
```

```python
# 1. Age
# 2. Weight
# 3. Blood Pressure
# 4. Cholesterol
data = np.array(data)
data
```

```python
# Part 1
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
min_vals = np.min(data, axis=0)
max_vals = np.max(data, axis=0)
print(f'Mean = {mean}')
print(f'Std = {std}')
print(f'Min = {min_vals}')
print(f'Max = {max_vals}')
```

```python
# Part 2
bp = data[ : ,2]# column 3
chol = data[ : ,3]# column 4
risk_patients = np.where((bp > 140) & (chol > 200))[0]
print(risk_patients)
```

```python
# Part 3
data_norm = (data - min_vals) / (max_vals - min_vals)
print(f'Data Normalization = \n\n{data_norm}')
```

```python
# Part 4
corr_matrix = np.corrcoef(data, rowvar=False)
print(f'Correlation Matrix = \n\n{corr_matrix}')
```

```python
# Part 5
weighted_norm = data[:,1]
bp_norm = data[:,2]
chol_norm = data[:,3]
risk_score = 0.3 * weighted_norm + 0.4 * bp_norm + 0.3 * chol_norm
print(f'Risk Score = {risk_score}')
```

```python
top3 = np.argsort(risk_score)[-3:][::-1] # [start : end : incerement]
```
